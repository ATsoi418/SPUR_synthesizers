```python
import csv
import pandas as pd
```

```python
with open('MMR.xlsx', 'rb') as mmr:
    mmr_df = pd.read_excel(mmr)
with open("Models by Manufacturer Resurgence 4-04-19 CA.xlsx", "rb") as m:
    mmr_ca_df = pd.read_excel(m)
```

```python
mmr_df
```

```python
cols = ['Firm','Model name','Sources', 'Sampler/Ropler',
 'Analog modeling',
 'Software synth',
 'DCOs',
 'Analog Filters',
 'Poly=1',
 'includes keyboard',
 'has knobs',
 'Memory',
 '# Ads']
mmr_ca_df = mmr_ca_df[cols].fillna(0)
mmr_ca_df.dropna(inplace = True) 
  
# new data frame with split value columns 
new = mmr_ca_df["Sources"].str.split("\n", n = 1, expand = True) 
  
# making seperate first name column from new data frame 
mmr_ca_df["Source_1"] = new[0] 
  
# making seperate last name column from new data frame 
mmr_ca_df['Source_2'] = new[1] 
```

```python
mmr_ca_df = mmr_ca_df[['Source_1','Sampler/Ropler',
 'Analog modeling',
 'Software synth',
 'DCOs',
 'Analog Filters',
 'Poly=1',
 'includes keyboard',
 'has knobs',
 'Memory',
 '# Ads']]
mmr_ca_df
```

```python
new_df = pd.merge(mmr_df, mmr_ca_df, on=['Source_1'], how='left')
```

```python
new_df.to_excel('MMR.xlsx', header=True, index=False)
```
