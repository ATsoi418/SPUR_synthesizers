```python
import pandas as pd
import requests
import json
from bs4 import BeautifulSoup
import urllib.request
import re
import csv
from PIL import Image
from io import BytesIO
```

```python
data = pd.read_csv('Edited_Coded_Models_by_manufacturer_resurgence.csv')
col1 = data['firm'].tolist()
col2 = data['model_name'].tolist()
col3 = data['url'].tolist()

brand = col1[:765]
mod_name = col2[:765]
mod_urls = col3[:765]
```

```python
print('brand: ', len(brand), '---', brand[:2], brand[763:])
print('mod_name: ', len(mod_name), '---', mod_name[:2], mod_name[763:])
print('mod_url: ', len(mod_urls), '---', mod_urls[:2], mod_urls[763:])
#print(len(mod_urls)) #should equal to 765
```

```python
pattern = re.compile(r'(www\.vintagesynth\.com\\\/sites\\\/default\\\/files(\\\/\w+)+\\\/2017-05\\\/(\w+)\.jpg)', re.MULTILINE | re.DOTALL)
```

```python
image_urls = []
for mod_url in mod_urls:
    script = BeautifulSoup(requests.get(mod_url).text, 'html.parser').find('script', text=pattern)
    if script:
        match = pattern.search(script.text)
        if match:
            result = match.group(1)
            url = result.replace("\\", "") 
            image_urls.append(url)
        
print(image_urls[:10])
```

```python
for index, image_url in enumerate(image_urls):
    try:
        mod = str(mod_name[index].replace("/","").replace(" ",""))
        print(str('http://'+image_url))
        urllib.request.urlretrieve(str('http://'+image_url), str('images/'+brand[index]+'_'+mod+'.jpg'))
        print('--- ', index, ': ', str('images/'+brand[index]+'_'+mod+'.jpg'), ' successfully inserted ---')
    except Exception as e:
        print(e)
```
