```python
import pandas as pd
import requests
import json
from bs4 import BeautifulSoup
import re
import csv
from PIL import Image
from io import BytesIO
```

```python
data = pd.read_csv('Edited_Coded_Models_by_manufacturer_resurgence.csv')
col = data['url'].tolist()
mod_urls = col[:765]
```

```python
print(mod_urls[764:])
print(len(mod_urls)) #should equal to 765
```

```python
pattern = re.compile(r'(www\.vintagesynth\.com\\\/sites\\\/default\\\/files(\\\/\w+)+\\\/2017-05\\\/(\w+)\.jpg)', re.MULTILINE | re.DOTALL)
```

,"image":{"@type":"ImageObject","url":"http:\/\/www.vintagesynth.com\/sites\/default\/files\/styles\/750x435sc\/public\/2017-05\/access_virus.jpg?itok=eJN--ft9","height":"435","width":"750"}}</script>

```python
image_urls = []
mod_codes = []
for mod_url in mod_urls:
    script = BeautifulSoup(requests.get(mod_url).text, 'html.parser').find('script', text=pattern)
    if script:
        match = pattern.search(script.text)
        if match:
            image_urls.append(match.group(1))
            mod_codes.append(match.group(3))
        
print(image_urls[:10])
print(mod_codes[:10])
```

```python
print(str('images/'+mod_codes[0]+'.jpg'))
```

```python
for index, image_url in enumerate(image_urls):
    try:
        img = Image.open(BytesIO(requests.get(str('http://'+image_url)).content))
        img.save(str('images/'+mod_codes[index]+'.jpg'))
        print('--- ', index, ': ', str('images/'+mod_codes[index]+'.jpg'), ' successfully inserted ---')
    except Exception as e:
        print(e)
```
